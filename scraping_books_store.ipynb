## Creating function for product link


```python
def prod_link(your_soup,book_no):
    link=your_soup[book_no].find('a')['href']
    book_link='https://books.toscrape.com/'+link
    return book_link
```

## Creating Function for product price


```python
def prod_price(your_soup,book_no):
    price=your_soup[book_no].find('p',class_='price_color')
    product_price=price.text.strip()
    return product_price[1:]
```

# creating a pandas dataframe

```python
column_name=['Name of book','Price','Availability','Link']
```

```python
import pandas as pd
```

```python
book_df=pd.DataFrame(columns=column_name)
```

```python
book_df
```

```python
for i in range(1,3):
    page_url=f'https://books.toscrape.com/catalogue/page-{i}.html'
    for j in range(1,19):
        list=[]
        web_page=requests.get(page_url)
        soup=BeautifulSoup(web_page.content, 'html.parser' )
        all_books_page=soup.find_all('li',class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')
        name=prod_name(all_books_page,j)
        stock=prod_stock(all_books_page,j)
        link=prod_link(all_books_page,j)
        price=prod_price(all_books_page,j)
        list.extend([name,price,stock,link])
        
        book_df.loc[len(book_df)]=list
    
```
